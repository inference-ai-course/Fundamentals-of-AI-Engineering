## Why environment checks matter

A Python environment is an isolated place where Python and installed packages live. Isolation prevents one project from breaking another project that needs different package versions.

For Week 3, the most important rule is: the terminal, `pip`, and the Jupyter kernel should all point to the same environment.


## Exercise 1: Check the Python executable

This cell prints the Python executable that is running notebook code. If it points to a system Python when you expected a project environment, select the correct Jupyter kernel before continuing.


```python
import sys
from pathlib import Path

print("Python version:", sys.version.split()[0])
print("Python executable:", sys.executable)
print("Current working directory:", Path.cwd())
```

## Exercise 2: Check pip through the active Python

Prefer `python -m pip` over plain `pip`. It guarantees that pip runs through the Python interpreter shown above.


```python
import subprocess
import sys

result = subprocess.run(
    [sys.executable, "-m", "pip", "--version"],
    text=True,
    capture_output=True,
    check=False,
)
print(result.stdout.strip() or result.stderr.strip())
```

## Exercise 3: Verify pandas

Week 3 depends on pandas for CSV loading and profiling. This import must work before you start Part 02.


```python
try:
    import pandas as pd
except ImportError as exc:
    raise RuntimeError(
        "pandas is not installed in this notebook kernel. Activate the Week 2 environment "
        "or select the correct Jupyter kernel, then try again."
    ) from exc

print("pandas version:", pd.__version__)
```

## Exercise 4: Understand the Jupyter kernel boundary

A notebook is a document. A kernel is the Python process that executes cells. Changing your terminal environment does not automatically change the notebook kernel.

If imports fail in a notebook but work in a terminal, compare:

- `sys.executable` in the notebook
- `which python` or `where python` in the terminal


```python
import os
import sys

print("Notebook kernel Python:")
print(sys.executable)

print("
PATH starts with:")
for item in os.environ.get("PATH", "").split(os.pathsep)[:5]:
    print("-", item)
```

## Exercise 5: Record dependencies

A dependency file helps another person rebuild the environment. `pip freeze` records installed package versions. For this course, it is enough evidence that your setup can be reproduced.


```python
import subprocess
import sys

result = subprocess.run(
    [sys.executable, "-m", "pip", "freeze"],
    text=True,
    capture_output=True,
    check=False,
)
packages = [line for line in result.stdout.splitlines() if line.strip()]
print("Installed package count:", len(packages))
print("First 10 packages:")
for line in packages[:10]:
    print("-", line)
```
