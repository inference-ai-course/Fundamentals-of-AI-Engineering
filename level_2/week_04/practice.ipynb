## Citation formatting template

Keep citations traceable: include `doc_id` and a short snippet.


```python
def format_citations(hits):
    out = []
    for h in hits:
        out.append({
            "doc_id": h.get("doc_id"),
            "chunk_id": h.get("chunk_id"),
            "snippet": (h.get("text") or "")[:200],
        })
    return out
```
