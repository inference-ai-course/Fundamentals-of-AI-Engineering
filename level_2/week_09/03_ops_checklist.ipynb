## Overview

An ops checklist is your runbook.

It should answer:

- what can fail
- what to monitor
- how to debug
- how to control cost

## Underlying theory: error budgets and leading indicators

If you set an SLO like “99% success”, you are implicitly allowing a 1% error budget.

Treat budget burn as a signal:

- if you are burning budget fast, stabilize before shipping changes

Leading indicators for RAG systems:

- retrieval hit rate drops
- refusal rate spikes (often indicates index/filters broken)
- context size spikes (often indicates runaway packing)
- 429/timeout rates spike (often indicates retry storms)

## Practice Steps

- Draft a checklist (monitoring + debug steps).
- Add explicit cost controls (tokens/context/rate limits).

Add a debug sequence for retrieval misses.


```python
DEBUG_RETRIEVAL_MISS = [
    "Find request_id from the failing response",
    "Inspect logs by component: retrieve -> generate -> validate",
    "Check retrieval returned any hits",
    "If hits empty for known in-KB questions: suspect ingestion/indexing/filters",
    "If hits exist but answer is wrong: suspect context assembly (ordering/truncation)",
    "If answer ok but citations missing: suspect citation enforcement/validation",
]

COST_CONTROLS = [
    "token caps",
    "context caps",
    "rate limits",
    "bounded retries with backoff",
]

print("debug_retrieval_miss:")
for s in DEBUG_RETRIEVAL_MISS:
    print("-", s)

print("cost_controls:")
for s in COST_CONTROLS:
    print("-", s)
```
